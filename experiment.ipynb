```python
# Import requests library
import requests

# Define the OCR API endpoint
url = "https://api.ocr.space/parse/image"

# Define the API key and the language
api_key = "K88232854988957"
language = "eng"

# Define the image file path
image_file = r"C:\Users\Ayoo\Desktop\webapp\predictions\imgs\20230508_122035.jpg"

# Open the image file as binary
with open(image_file, "rb") as f:
    # Define the payload for the API request
    payload = {
        "apikey": api_key,
        "language": language,
        "isOverlayRequired": True, # Optional, set to True if you want the coordinates of the words
        "OCREngine": 2 # OCR Engine 2 for Layoutlmv3
    }
    # Define the file parameter for the API request
    file = {
        "file": f
    }
    # Send the POST request to the OCR API
    response = requests.post(url, data=payload, files=file)

# Check the status code of the response
if response.status_code == 200:
    # Parse the JSON response
    result = response.json()
    # Print the parsed text
    print(result)
else:
    # Print the error message
    print("Error: " + response.text)
```
